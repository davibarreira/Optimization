```julia
function CC(S)
    T1 = (S.-S[:,1])[:,2:end]
    T2 = pinv(T1' * T1) # MODIFIED THE ORIGINAL BY THE PSEUDO-INVERSE INSTEAD OF THE INVERSE. IT SEEMS TO WORK... WHY?
    T3 = diag((S.-S[:,1])[:,2:end]' * (S.-S[:,1])[:,2:end])
    cc = S[:,1] + 0.5.*T1 * T2 * T3
    return cc
end
```

```julia
f1 = [1,40]
f2 = [20,0]
f3 = [-10,-10]
f4 = [-10,10]

S = [f1 f2 f3 f4]
scatter(S'[:,1],S'[:,2])
fc = CC(S)
scatter!([fc[1]],[fc[2]])
```
